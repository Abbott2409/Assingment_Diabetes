```python
# find out any duplicate values by keeping the first row same
data = data.drop_duplicates(keep='first')
```

```python
data
```

```python
#check the mean of the value depending on their category (i.e 0 or 1)

data.groupby('Outcome').mean()
```

```python
#the difference between mean and medium is a indicator of how much skewed is the data is
data.groupby('Outcome').agg(['mean','median'])
```

```python
for i,col in enumerate(data.columns[:-1]):
    plt.figure(i)
    sns.distplot(data[col]);
```

```python
#scatter matrix: 
sns.pairplot(data, hue="Outcome", diag_kind = 'hist');
```

```python
#corelation  Quantify which variable/attributes are corrleated with each others
data.corr()
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(np.abs(data.corr()), annot=True, cmap='viridis', fmt='0.2f');
```

```python
#box plot
sns.boxplot(x="Outcome",y="BMI",data=data)
```
